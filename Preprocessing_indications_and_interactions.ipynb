```python

import re

def get_side_effects(text,start_text):
  start_index = text.find(str(start_text))+len(start_text)

  cleaned_text = text[start_index:].strip()
  desired_text = cleaned_text = re.sub(r'\s*\(.*?\)\s*', ' ', cleaned_text)

  # Split the text into separate lines and filter out empty lines
  side_effects = [line.strip() for line in desired_text.splitlines() if line.strip()]

  return side_effects

import re
def get_uses(text):

  cleaned_text=text.replace('\n', '').strip()
  cleaned_text=cleaned_text.replace('\\n', '').strip()
  cleaned_text=cleaned_text.replace('\\', '').strip()
  cleaned_text=cleaned_text.replace('ul', '').strip()
  cleaned_text=cleaned_text.replace('"', '').strip()
  cleaned_text = re.sub(r'\s*\(.*?\)\s*', ' ', cleaned_text)

  # Split the string by numbers
  split_result = re.split(r'\d+', cleaned_text)
  # Remove empty strings and strip whitespace
  uses= [part.strip() for part in split_result if part.strip()]

  return uses

def get_components(text):
  components = [part.strip() for part in text.split('+')]

  return components

df = pd.DataFrame(data)
# Apply the cut_text function to all interactions
for i in range(len(data)):
    df['interactions'][i] = get_side_effects(str(data['interactions'][i]),str("Common side effects of"))
    df['indications'][i]=get_uses(str(data['indications'][i]))
    df['contains'][i]=get_components(str(data['contains'][i]))
    print(i)

#df.to_excel('/content/drive/MyDrive/MID/dataset/indications_and_interactions_by_therapeutic.xlsx', index=False)
df.head()
```

```python
# Remove the first item from each list in the 'interactions' column
df['interactions'] = df['interactions'].apply(lambda x: x[1:] if len(x) > 0 else x)

print(df)
```

**Dictionary Indications**

```python
# Flatten the data to create a structured DataFrame
import ast
f=df['indications']
index=1
t=0
upperframe= []
for sublist in f:
    sublist = ast.literal_eval(sublist)
    print(t)
    t=t+1
    for indication in sublist:
        upperframe.append((index,indication))
        index=index+1

# Create DataFrame
d = pd.DataFrame(upperframe, columns=['Index', 'Indication'])

# Remove duplicates based on the 'Indication' column
d = d.drop_duplicates(subset=['Indication'])

# Reset the index and create a new 'Index' column starting from 1
d = d.reset_index(drop=True)  # Reset index
d['Index'] = d.index + 1  # Create a new 'Index' column starting from 1

# Export to Excel
d.to_excel('/content/drive/MyDrive/MID/dataset/Dictionar_indications.xlsx', index=False)

print("Data exported to conditions_treatments.xlsx successfully.")

print(d)
```
